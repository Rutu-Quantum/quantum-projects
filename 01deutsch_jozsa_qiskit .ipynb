# DeutschвҖ“Jozsa Algorithm using Qiskit

## Objective
-To determine whether a Boolean function is      constant or
 balanced using a quantum algorithm, demonstrating quantum computational advantage.

## Theory

-The Deutsch-Jozsa algorithm is one of the earliest examples of a quantum algorithm
 that demonstrates a clear advantage over classical computation.

A Boolean function f(x) is:
- Constant: same output for all inputs  
- Balanced: returns 0 for half inputs and 1 for half  

Classically, multiple evaluations are needed. Quantum mechanically, the answer is
obtained in a single evaluation using superposition and interference.

## Methodology

1. Initialize qubits  
2. Apply Hadamard gates to create superposition  
3. Apply oracle representing the function  
4. Apply Hadamard again  
5. Measure the output  

The measurement result determines whether the function is constant or balanced.

```python

!pip install qiskit qiskit-aer --quiet


#Deutsch-Jozsa Algorithm using Qiskit

#import libraries
from qiskit import QuantumCircuit,transpile
from qiskit_aer import AerSimulator
import numpy as np

# oracle
def dj_oracle(n, oracle_type='balanced'):
    oracle = QuantumCircuit(n+1)

    if oracle_type == 'constant':
        # return either always 0 or always 1
        if np.random.randint(2) == 1:
            oracle.x(n)

    elif oracle_type == 'balanced':
        # Apply CNOT from each input qubit to output qubit
        for qubit in range(n):
            oracle.cx(qubit, n)

    return oracle


# number of input qubits
n = 5

# Create quantum circuit
qc = QuantumCircuit(n+1, n)

# Step 1: Initialize output qubit to |1>
qc.x(n)

# Step 2: Apply Hadamard to all qubits
qc.h(range(n+1))

# Step 3: Apply oracle
oracle = dj_oracle(n, oracle_type='balanced')  # change to 'constant' to test
qc.append(oracle, range(n+1))

# Step 4: Apply Hadamard to input qubits
qc.h(range(n))

# Step 5: Measure input qubits
qc.measure(range(n), range(n))


# Draw circuit (text)
print(qc.draw())

# Simulate
sim = AerSimulator()

compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

# Get counts
counts = result.get_counts()

print("Measurement Results:", counts)
# Interpretation
if list(counts.keys()) == ['0'*n]:
    print("Function is CONSTANT")
else:
    print("Function is BALANCED")
```
