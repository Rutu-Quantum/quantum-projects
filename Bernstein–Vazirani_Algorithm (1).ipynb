# BernsteinвҖ“Vazirani Algorithm

## Objective
To determine a hidden binary string using a quantum algorithm with a single query.

## Theory
The BernsteinвҖ“Vazirani algorithm efficiently finds a hidden string encoded in a function. Classically, multiple queries are required, whereas quantum computation solves it in one step.

```python
!pip install qiskit qiskit-aer --quiet

#BernsteinвҖ“Vazirani Algorithm

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# secret string
s = "10101"
n = len(s)

# circuit
qc = QuantumCircuit(n+1, n)

# Step 1: Initialize
qc.x(n)
qc.h(range(n+1))

# Step 2: Oracle
for i in range(n):
    if s[i] == '1':
        qc.cx(i, n)

# Step 3: Hadamard on input qubits
qc.h(range(n))

# Step 4: Measure
qc.measure(range(n), range(n))

print(qc.draw())

# Simulate
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

counts = result.get_counts()
print("Measurement Results:", counts)
```

## Result
The measured output directly reveals the hidden binary string.
