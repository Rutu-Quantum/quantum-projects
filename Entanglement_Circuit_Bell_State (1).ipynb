# Quantum Entanglement

## Objective
To create and observe entangled quantum states.

## Theory
Entanglement is a fundamental quantum phenomenon where qubits become correlated such that the state of one affects the other.

```python
!pip install qiskit qiskit-aer --quiet

#Entanglement Circuit (Bell State)

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2, 2)

# Create entanglement
qc.h(0)
qc.cx(0, 1)

# Measure
qc.measure([0,1], [0,1])

print(qc.draw())

# Simulate
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

counts = result.get_counts()
print("Measurement Results:", counts)
```

## Result
The measurement results show correlated outcomes, confirming entanglement.
