# Grover's Algorithm

## Objective
To search for a marked element in an unsorted database with quadratic speedup.

## Theory
GroverвҖҷs algorithm amplifies the probability of the correct solution using repeated applications of oracle and diffusion operators.

```python
!pip install qiskit qiskit-aer --quiet
#GroverвҖҷs Algorithm (2-qubit)

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# number of qubits
n = 2

qc = QuantumCircuit(n, n)

# Step 1: Superposition
qc.h(range(n))

# Step 2: Oracle (mark |11>)
qc.cz(0, 1)

# Step 3: Diffusion operator
qc.h(range(n))
qc.x(range(n))
qc.h(1)
qc.cx(0,1)
qc.h(1)
qc.x(range(n))
qc.h(range(n))

# Step 4: Measure
qc.measure(range(n), range(n))

print(qc.draw())

# Simulate
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

counts = result.get_counts()
print("Measurement Results:", counts)
```

## Result
The marked state appears with highest probability after amplification.
